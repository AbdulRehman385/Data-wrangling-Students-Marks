```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# In this task we are supposed to find average marks of students
# Secondly we are supposed to provide grades on such marks
```

```python
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 90, 78, 88, 92],
    'Science': [89, 85, 82, 91, 88],
    'English': [90, 88, 79, 87, 93]
})
df.head()
```

```python
# Now create a column named Average using apply function
df['Average'] = df[['Math', 'Science', 'English']].apply(lambda x: np.mean(x), axis = 1)
```

```python
df.head()
```

```python
# Lets round off to make it more clear
df['Average'] = df['Average'].apply(lambda x: round(x,2))
```

```python
df.head()
```

```python
# Create a function for grades
def grades(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80 and marks < 90:
        return 'B'
    elif marks >= 70 and marks < 80:
        return 'C'
    else:
        return 'D'
```

```python
df['Grades'] = df['Average'].apply(grades)
```

```python
df.head()
```
